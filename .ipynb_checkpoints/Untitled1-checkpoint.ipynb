```python
from PyPDF2 import PdfFileReader , PdfFileWriter
from pathlib import Path

import spacy
import re
from spacy.language import Language
```

```python
pdf = PdfFileReader('Physics/55-1-1 Physics.pdf')
pdf.documentInfo
```

```python
pdf.getNumPages()
```

```python
page_1_object = pdf.getPage(10) # 0 is for page no 1
print(page_1_object)
```

```python
page_1_object.extractText()
```

```python
page_text = ''

for i in range(2,pdf.getNumPages(),2):
    print(i)
    page_object = pdf.getPage(i)
    text_1= page_object.extractText()
    page_text = page_text+text_1
```

```python
page_text
```

```python
text = re.sub(r'\n|\\n','',page_text)
text
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
for sent in doc.sents:
    print(sent,'\n')
```

```python
pattern = r'\d'
pattern2 = r'\.'

for token in doc:
        if re.match(pattern , token.text) and re.match(pattern2 , doc[token.i+1].text):
            doc[token.i + 2].is_sent_start = True
```
